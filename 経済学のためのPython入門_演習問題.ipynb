```python
# 演習４　⑵
def disposable_income(income, tax):
    """可処分所得を計算する
    引数 :
     income: 所得
     tax: 所得税
    戻り値 :
     可処分所得"""
    return income - tax
```

```python
# 演習４　⑶
result = disposable_income(600, 130)
print(f"可処分所得は{result}万円です。")
```

```python
# 演習５　⑴
def present_value(future_value, rate=0.03, years=10):
    pv = future_value / (1 + rate) ** years
    """キャッシュフローを評価する際の現在価値(PV)を計算する。
    引数 :
     future_value: 将来受け取る金額
     rate: 割引率（デフォルトは0.03=3%）
     years: 何年後に受け取るか（デフォルトは10）
  　戻り値 :
     現在価値"""
    return pv
```

```python
# 演習５　⑵
result = present_value(100_000_000)
print(f"10年後の１億円の現在価値は{result:,.0f}円です。")
```
