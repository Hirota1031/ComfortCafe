```python
def demand(p):
  return 40 / p ** (0.1)

def supply(p):
  return p ** (1.5)

def d_larger_than_s(p):

  d = demand(p)
  s = supply(p)

  print(f"価格={p}")
  print(f"需要={d:.2f}")
  print(f"供給={s:.2f}")

  return d > s

d_larger_than_s(9)
```

```python
def future_value_docstring(x, r=0.03, t=10):
    """ 投資の将来価値を計算する
    引数 :
     x(float): 初期投資額
     r(float): 期待収益率（デフォルトは3%）
     t(int): 投資期間（デフォルトは10年）
    戻り値 :
     float: t年後の収益"""

    result = x * (1 + r)**t

    return result
```
