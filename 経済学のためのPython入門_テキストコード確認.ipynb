```python
def demand(p):
  return 40 / p ** (0.1)

def supply(p):
  return p ** (1.5)

def d_larger_than_s(p):

  d = demand(p)
  s = supply(p)

  print(f"価格={p}")
  print(f"需要={d:.2f}")
  print(f"供給={s:.2f}")

  return d > s

d_larger_than_s(9)
```

```python
def future_value_docstring(x, r=0.03, t=10):
    """ 投資の将来価値を計算する
    引数 :
     x(float): 初期投資額
     r(float): 期待収益率（デフォルトは3%）
     t(int): 投資期間（デフォルトは10年）
    戻り値 :
     float: t年後の収益"""

    result = x * (1 + r)**t

    return result
```

```python
def wealth_years(wealth=100,
                 target=200,
                 r=0.03):
    """
    引数 :
        wealth: 初期資産
        target: 目標資産
        r: 年間収益率
    戻り値 :
        目標資産額以上の達成に必要な年数（表示あり）"""
    year = 0

    while wealth < target:
        wealth *= 1+ r
        year += 1

    print(f"必要年数:{year}年")
    return year
```

```python
n = 1_000
x2 = [random.gauss() for _ in range(n)]
u2 = [random.gauss() for _ in range(n)]
y2 = [i + j for i, j in zip(x2, u2)]

plt.scatter(x2, y2)
plt.show()
```

```python
x3 = [random.gauss() for _ in range(n)]
u3 = [random.gauss() for _ in range(n)]
y3 = [-i + 0.5*j for i, j in zip(x3, u3)]

plt.scatter(x2, y2,
            marker = "o",
            s = 50,
            color = "gray",
            alpha = 0.8,
            label = "(x2,y2)")
plt.scatter(x3, y3,
            marker = "x",
            s = 50,
            color = "black",
            label = "(x3,y3)")

plt.title("散布図", size = 18)
plt.xlabel("x2, x3", size = 15)
plt.ylabel("x3, y3", size = 15)
plt.legend(fontsize = 10)
plt.show()
```
